# Criação de Nuvem de Tags com Python. 

Faz parte do meu estudo sobre Ciência de Dados e Python.

As Nuvens de Tags podem ser bem interessantes para avliação de textos médios ou longos sem, necessariamente, precisar ler. Pode ser utilizado para se ter uma ideia geral de um discurso, artigo ou notícia, por exemplo. 


# Stopwords em Português

```
import nltk
from nltk.corpus import stopwords
```

```
nltk.download('stopwords')
print(stopwords.words('portuguese')) 
```

# Dividindo o texto em sentenças

```
from nltk.tokenize import word_tokenize 

import nltk
nltk.download('punkt')
```

# Texto com e sem stopwords

```
#só para colocar um cor no texto
import termcolor
from termcolor import colored
texto = input("Coloque eu texto aqui")
  
stop_words = set(stopwords.words('portuguese'))

word_tokens = word_tokenize(texto)

texto_filtrado = [w for w in word_tokens if not w in stop_words]

texto_filtrado = []

for w in word_tokens:
    if w not in stop_words:
        texto_filtrado.append(w)

print colored("Essas são as palavras sem retirar as stopwords:", "green", word_tokens)
print("Essas são as palavras depois de retirar as stopwords:", texto_filtrado)
```

# Texto completo dividido em sentenças e sem stopwords

```
texto_completo = ', '.join(texto_filtrado)
print(texto_completo)
```

# Criando e printando a nuvem de tags de acordo com os parâmetros

```
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
```

```
stops = stopwords.words("portuguese")
mapa_cores = ListedColormap(["orange", "green", "red", "magenta"])
nuvem = WordCloud(background_color = "white",  height=600, width=1200, colormap = mapa_cores, stopwords = stops, max_words = 50)
```

```
nuvem.generate(texto_completo)
```

```
from matplotlib import pyplot as plt
plt.figure( figsize=(10,5))
plt.imshow(nuvem)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
